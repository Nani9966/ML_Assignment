# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear_reg=LinearRegression()
```

```python
linear_reg.fit(X_train,y_train)
```

```python
linear_reg_pred=linear_reg.predict(X_test)
```

```python
linear_reg_pred
```

```python
from sklearn.metrics import r2_score
r2score_linear=r2_score(y_test, linear_reg_pred)
print("Our Linear Regression model has {} % accuracy".format(round(r2score_linear*100,3)))
```

```python

adjusted_r2_score_linear_reg=1-((1-r2score_linear)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
print("Adjusted R square accuracy is {} % ".format(round(adjusted_r2_score_linear_reg*100,3)))
```
